## Q4. Gaussian Mixture Model (GMM): Full EM + Plot (25 marks)

Implement a full **EM algorithm** for a **1D two-component** Gaussian Mixture and add a plot of:
- the data histogram, and
- the learned mixture density overlaid as $\sum_k \pi_k\, \mathcal{N}(x\mid \mu_k,\sigma_k^2)$.

**Implement**
1. `fit_gmm_em_1d(x, K=2, max_iter=200, tol=1e-5)` → `(pi, mu, sigma2, loglik_trace)`
   - Randomly initialise parameters (or via simple K-means-like init)
   - E-step: responsibilities $\gamma_{nk}$
   - M-step: update $\pi_k$, $\mu_k$, $\sigma_k^2$ (use variance floor `EPS`)
   - Track log-likelihood and stop when improvement < `tol`.
2. `plot_gmm_1d(x, pi, mu, sigma2)` to visualise histogram and mixture curve.

**Marking (25)**
- Correct E-step and M-step (12)
- Monotonic log-likelihood and convergence (6)
- Clear plotting function (5)
- Code clarity and comments (2)

```python
def normal_pdf_1d(x: np.ndarray, mu: float, sigma2: float) -> np.ndarray:
    """1D Gaussian PDF evaluated element-wise for array x.
    Returns an array of same shape as x.
    """
    sigma2 = max(sigma2, EPS)
    return (1.0/np.sqrt(2*np.pi*sigma2)) * np.exp(-0.5*((x - mu)**2)/sigma2)


def fit_gmm_em_1d(x: np.ndarray, K: int = 2, max_iter: int = 200, tol: float = 1e-5):
    """Fit a 1D K-component GMM using EM.
    Parameters
    ----------
    x : array-like of shape (n,)
        1D data samples.
    K : int
        Number of mixture components (default 2).
    max_iter : int
        Maximum EM iterations.
    tol : float
        Convergence tolerance on log-likelihood improvement.
    Returns
    -------
    pi : array of shape (K,)
        Mixing coefficients (sum to 1).
    mu : array of shape (K,)
        Component means.
    sigma2 : array of shape (K,)
        Component variances (floored at EPS).
    loglik_trace : list[float]
        Log-likelihood values per iteration.
    """
    x = np.asarray(x).ravel()
    n = x.shape[0]

    # --- Initialisation ---
    # Simple init: choose K quantiles as initial means, equal variances, equal mixing
    qs = np.linspace(0.1, 0.9, K)
    mu = np.quantile(x, qs)
    sigma2 = np.full(K, np.var(x) + 1e-2)
    pi = np.full(K, 1.0/K)

    loglik_trace = []

    for it in range(max_iter):
        # --- E-step: responsibilities gamma[n,k] ---
        # Compute weighted densities for each component
        num = np.stack([pi[k] * normal_pdf_1d(x, mu[k], sigma2[k]) for k in range(K)], axis=1)  # (n,K)
        den = np.sum(num, axis=1, keepdims=True) + EPS
        gamma = num / den  # (n,K)

        # --- M-step: update parameters ---
        Nk = gamma.sum(axis=0) + EPS  # effective counts (K,)
        pi = Nk / n
        mu = (gamma * x[:, None]).sum(axis=0) / Nk
        # Unbiased-ish variance update per component
        sigma2 = (gamma * (x[:, None] - mu[None, :])**2).sum(axis=0) / Nk
        sigma2 = np.maximum(sigma2, EPS)

        # --- Log-likelihood ---
        ll = np.sum(np.log(np.sum(np.stack([pi[k] * normal_pdf_1d(x, mu[k], sigma2[k]) for k in range(K)], axis=1), axis=1) + EPS))
        loglik_trace.append(float(ll))

        # --- Convergence check ---
        if it > 0 and (loglik_trace[-1] - loglik_trace[-2]) < tol:
            break

    return pi, mu, sigma2, loglik_trace


def plot_gmm_1d(x: np.ndarray, pi: np.ndarray, mu: np.ndarray, sigma2: np.ndarray, bins: int = 30):
    """Plot the data histogram and the learned GMM density curve.
    Note: This function *plots*; it does not return anything.
    """
    x = np.asarray(x).ravel()
    xs = np.linspace(np.min(x)-1.0, np.max(x)+1.0, 400)
    # Mixture density
    mix_pdf = sum(pi[k]*normal_pdf_1d(xs, mu[k], sigma2[k]) for k in range(len(pi)))

    fig, ax = plt.subplots(figsize=(7,4))
    ax.hist(x, bins=bins, density=True, alpha=0.4, color='tab:gray', label='Data (hist)')
    ax.plot(xs, mix_pdf, lw=2.0, color='tab:red', label='GMM mixture density')
    for k in range(len(pi)):
        ax.plot(xs, pi[k]*normal_pdf_1d(xs, mu[k], sigma2[k]), '--', lw=1.5, label=f'Comp {k+1}')
    ax.set_xlabel('x')
    ax.set_ylabel('Density')
    ax.legend()
    ax.set_title('Learned 1D GMM')
    plt.show()
```
